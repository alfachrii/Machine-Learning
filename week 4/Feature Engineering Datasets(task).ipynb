```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df1 = pd.read_csv('/content/drive/MyDrive/machine learning/week 4/test.csv')
    df2 = pd.read_csv('/content/drive/MyDrive/machine learning/week 4/train.csv')
except FileNotFoundError:
    print("Error: One or both of the CSV files were not found.")
    exit()


def analyze_dataset(df, dataset_name):
    print(f"\n--- Analysis for {dataset_name} ---\n")
    print("Data Shape:", df.shape)
    print("\nData Types:\n", df.dtypes)
    print("\nDescriptive Statistics:\n", df.describe(include='all'))
    print("\nMissing Values:\n", df.isnull().sum())

    for col in df.columns:
        print(f"\n--- Feature: {col} ---")

        if pd.api.types.is_numeric_dtype(df[col]):
            # Numerical feature analysis
            print(f"Unique values: {df[col].nunique()}")
            plt.figure(figsize=(10, 6))
            sns.histplot(df[col], kde=True)
            plt.title(f"Distribution of {col} in {dataset_name}")
            plt.show()

            plt.figure(figsize=(10, 6))
            sns.boxplot(y=df[col])
            plt.title(f"Boxplot of {col} in {dataset_name}")
            plt.show()

        elif pd.api.types.is_string_dtype(df[col]) or pd.api.types.is_categorical_dtype(df[col]):
            # Categorical feature analysis
            print(f"Unique values: {df[col].unique()}")

            plt.figure(figsize=(10, 6))
            df[col].value_counts().plot(kind='bar')
            plt.title(f"Frequency of {col} in {dataset_name}")
            plt.xlabel(col)
            plt.ylabel("Frequency")
            plt.show()

            print(f"Value Counts:\n {df[col].value_counts()}")


        else:
            print("Unsupported data type for visualization.")


analyze_dataset(df1, "Dataset 1")
analyze_dataset(df2, "Dataset 2")
```

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y_true = df2['SalePrice'] # Example: Assuming 'SalePrice' is the target variable in df2
y_pred = np.random.rand(len(y_true)) * y_true.max() # Example: Random predictions

# Calculate metrics
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

# Create a matrix
evaluation_matrix = pd.DataFrame({
    'Metric': ['RMSE', 'MSE', 'R-squared'],
    'Value': [rmse, mse, r2]
})

evaluation_matrix
```
